```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate heart rate in beats per second and beats per minute
def calculate_heart_rate(ibi_seconds):
    if len(ibi_seconds) == 0:
        return 0, 0  # Return zero HR if no IBIs
    hr_bps = 1 / np.mean(ibi_seconds)
    hr_bpm = 60 * hr_bps
    return hr_bps, hr_bpm

# Function to plot heart rate data for a subject
def plot_subject_hr(subject_number, peak_timestamps):
    # Convert peak timestamps to datetime format
    peak_times = pd.to_datetime(peak_timestamps, unit='ns')

    # Filter out duplicate timestamps
    unique_timestamps = peak_times.drop_duplicates()

    # Calculate interbeat intervals (IBI) in seconds
    time_diff = np.diff(unique_timestamps).astype(int)  # Difference in nanoseconds
    ibi_seconds = time_diff / 1e9  # Convert nanoseconds to seconds

    # Print detected peaks and IBIs
    # print("Detected Peaks:", unique_timestamps)
    # print("Interbeat Intervals (IBIs):", ibi_seconds)

    # Calculate heart rate
    hr_bps, hr_bpm = calculate_heart_rate(ibi_seconds)
    print(f"Heart Rate (bps) for Subject {subject_number}:", hr_bps)
    print(f"Heart Rate (bpm) for Subject {subject_number}:", hr_bpm)

    # Create an intermediate DataFrame
    intermediate_df = pd.DataFrame({
        'timestamp': unique_timestamps[1:],  # Timestamps corresponding to the calculated IBIs
        'HR (bps)': [calculate_heart_rate([ibi])[0] for ibi in ibi_seconds],  # Calculate HR for each IBI
        'HR (bpm)': [calculate_heart_rate([ibi])[1] for ibi in ibi_seconds]   # Calculate HR for each IBI
    })

    # Format timestamp for plotting
    intermediate_df['timestamp_str'] = intermediate_df['timestamp'].dt.strftime('%H:%M:%S.%f')

    print(intermediate_df['HR (bps)'])

    # Plotting heart rate data
    plt.figure(figsize=(10, 6))
    plt.plot(intermediate_df['timestamp_str'], intermediate_df['HR (bps)'], label=f'Subject {subject_number} HR (bps)')
    plt.plot(intermediate_df['timestamp_str'], intermediate_df['HR (bpm)'], label=f'Subject {subject_number} HR (bpm)', linestyle='--')
    plt.xlabel('Time')
    plt.ylabel('Heart Rate')
    plt.title(f'Subject {subject_number} Heart Rate')
    plt.legend()
    plt.xticks(rotation=45)
    plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))  # Adjust number of ticks
    plt.show()

    return ibi_seconds

# Iterate over all subjects
for subject_number in range(2, 15):
    file_path_systolic_peaks = f'/content/drive/MyDrive/misoAR/csv_extracted/subject{subject_number}/systolic_peaks.csv'
    data_systolic_peaks = pd.read_csv(file_path_systolic_peaks)
    ibi_seconds = plot_subject_hr(subject_number, data_systolic_peaks['systolic_peak_timestamp'])
    print(f"Subject {subject_number} Calculated IBIs (in seconds):", ibi_seconds)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Function to calculate heart rate in beats per second and beats per minute
def calculate_heart_rate(ibi_seconds):
    if len(ibi_seconds) == 0:
        return 0, 0  # Return zero HR if no IBIs
    hr_bps = 1 / np.mean(ibi_seconds)
    hr_bpm = 60 * hr_bps
    return hr_bps, hr_bpm

# Function to plot BVP data for a subject and overlay detected peaks
def plot_subject_bvp_with_peaks(subject_number, offset):
    # Load BVP data from CSV file
    file_path_bvp = f'/content/drive/MyDrive/misoAR/csv_extracted/subject{subject_number}/bvp.csv'
    data_bvp = pd.read_csv(file_path_bvp)

    # Convert unix_timestamp column to datetime format
    data_bvp['timestamp'] = pd.to_datetime(data_bvp['unix_timestamp'], unit='ns')

    # Detect peaks in the BVP data
    peaks, _ = find_peaks(data_bvp['bvp'], height=0)  # Adjust the height parameter as needed

    # Plot BVP data with detected peaks
    plt.figure(figsize=(10, 6))
    plt.plot(data_bvp['timestamp'], data_bvp['bvp'], label=f'Subject {subject_number} BVP', alpha=0.7)
    plt.scatter(data_bvp['timestamp'].iloc[peaks], data_bvp['bvp'].iloc[peaks], color='red', marker='x', label='Detected Peaks')
    plt.xlabel('Time (ns)')  # Adjust the x-axis label to indicate nanoseconds
    plt.ylabel('BVP')
    plt.title(f'Subject {subject_number} BVP with Detected Peaks')
    plt.legend()
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.show()

# Plot
plot_subject_bvp_with_peaks(2, 0)
plot_subject_bvp_with_peaks(3, 0)
plot_subject_bvp_with_peaks(4, 0)
plot_subject_bvp_with_peaks(5, 0)
plot_subject_bvp_with_peaks(6, 0)
plot_subject_bvp_with_peaks(7, 0)
plot_subject_bvp_with_peaks(8, 0)
plot_subject_bvp_with_peaks(9, 0)
plot_subject_bvp_with_peaks(10, 0)
plot_subject_bvp_with_peaks(11, 0)
plot_subject_bvp_with_peaks(12, 0)
plot_subject_bvp_with_peaks(13, 0)
plot_subject_bvp_with_peaks(14, 0)
```

```python
# Save calculated heart rate to csv file

# Function to calculate heart rate in beats per second and beats per minute
def calculate_heart_rate(ibi_seconds):
    if len(ibi_seconds) == 0:
        return 0, 0  # Return zero HR if no IBIs
    hr_bps = 1 / np.mean(ibi_seconds)
    hr_bpm = 60 * hr_bps
    return hr_bps, hr_bpm

# Function to calculate heart rate variability (HRV) in seconds
def calculate_hrv(ibi_seconds):
    if len(ibi_seconds) < 2:
        return 0  # Return zero HRV if less than 2 IBIs
    hrv_seconds = np.std(ibi_seconds)
    return hrv_seconds

# Function to save HR and HRV to CSV
def save_hr_hrv_to_csv(subject_number, peak_timestamps):
    # Convert peak timestamps to datetime format
    peak_times = pd.to_datetime(peak_timestamps, unit='ns')

    # Filter out duplicate timestamps
    unique_timestamps = peak_times.drop_duplicates()

    # Calculate interbeat intervals (IBI) in seconds
    time_diff = np.diff(unique_timestamps).astype(int)  # Difference in nanoseconds
    ibi_seconds = time_diff / 1e9  # Convert nanoseconds to seconds

    # Calculate heart rate
    hr_bps, hr_bpm = calculate_heart_rate(ibi_seconds)

    # Calculate HRV
    hrv_seconds = calculate_hrv(ibi_seconds)

    # Create DataFrame with timestamp, HR (bps), HR (bpm), and HRV
    data = {
        'timestamp': unique_timestamps[1:],
        'HR (bps)': [hr_bps] * len(ibi_seconds),
        'HR (bpm)': [hr_bpm] * len(ibi_seconds),
        'HRV (seconds)': [hrv_seconds] * len(ibi_seconds)
    }
    df = pd.DataFrame(data)

    # Save DataFrame to CSV file
    csv_filename = f'/content/drive/MyDrive/misoAR/csv_extracted/subject{subject_number}/subject{subject_number}_hr_hrv.csv'
    df.to_csv(csv_filename, index=False)

    # Print confirmation
    print(f"HR and HRV data for Subject {subject_number} saved to {csv_filename}")

# Iterate over all subjects
for subject_number in range(2, 15):
    file_path_systolic_peaks = f'/content/drive/MyDrive/misoAR/csv_extracted/subject{subject_number}/systolic_peaks.csv'
    data_systolic_peaks = pd.read_csv(file_path_systolic_peaks)
    save_hr_hrv_to_csv(subject_number, data_systolic_peaks['systolic_peak_timestamp'])
```
