### Create Spaghetti Plot for each sensor

```python
import pandas as pd
import matplotlib.pyplot as plt

# BVP

# Define a function to plot BVP data for a given subject number
def plot_subject_bvp(subject_number, offset):
    # Load BVP data from CSV file
    file_path_bvp = f'/Users/sjtok/misophoniaAR/misoAR/Empatica_MisoAR_pilot/csv_extracted/subject{subject_number}/bvp.csv'
    data_bvp = pd.read_csv(file_path_bvp)

    # Plotting BVP data
    plt.plot(data_bvp['bvp'], label=f'Subject {subject_number}', alpha=0.7)
    plt.xticks([])  # Remove x-axis labels

# Plotting BVP data for subjects 2 through 14
plt.figure(figsize=(10, 6))
offset = 0
for subject_number in range(2, 15):
    plot_subject_bvp(subject_number, offset)
    offset += 1

# Adding title and labels
plt.title('Blood Volume Pulse Data Over Time')
plt.ylabel('BVP')
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define a function to plot skin temperature data for a given subject number
def plot_subject_temperature(subject_numbers):
    plt.figure(figsize=(10, 6))
    
    for subject_number in subject_numbers:
        # Load temperature data from CSV file
        file_path_temperature = f'/Users/sjtok/misophoniaAR/misoAR/Empatica_MisoAR_pilot/csv_extracted/subject{subject_number}/temperature.csv'
        data_temperature = pd.read_csv(file_path_temperature)

        # Calculate time elapsed (in minutes) using the UNIX timestamp
        start_time = data_temperature['unix_timestamp'].iloc[0]  # Get the first timestamp
        data_temperature['time_elapsed'] = (data_temperature['unix_timestamp'] - start_time) / (60 * 1e9)  # Convert nanoseconds to minutes

        # Plotting temperature data
        plt.plot(data_temperature['time_elapsed'], data_temperature['temperature'], label=f'Subject {subject_number}', alpha=0.7)

    # Adding title and labels
    plt.title('Skin Temperature Over Time')
    plt.xlabel('Time Elapsed (minutes)')
    plt.ylabel('Temperature')
    plt.legend()

    # Display the plot
    plt.grid(True)
    plt.tight_layout()
    plt.show()


plot_subject_temperature([13])

plot_subject_temperature([14])

```
